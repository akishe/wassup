# Q2-1

아래의 파일을 이용하여 데이타프레임을 생성하여라.<br>
   
전국 평균 분양가격.csv 

````python
df = pd.read_csv('./data/전국 평균 분양가격.csv',encoding = 'cp949' )
````

````python
df.head()
````

# Q2-2

행과 열을 아래의 명령어를 이용하여 서로 변경하여라.  (df는 데이타프레임명)
```
df = df.T

```

````python
df = df.T
df.head()
````

# Q2-3

1행을 삭제하고 2행을 컬럼으로 지정하여라. 

````python
df.columns = df.iloc[0]
````

````python
df = df.drop(index='지역',axis=1)
````

````python
df = df.reset_index()
df = df.rename(columns = {'index':'지역'})
````

````python
df['지역'] = df['지역'].apply(lambda x: x.split('년')[0] +'-'+ x.split('년')[1])
df['지역'] = df['지역'].apply(lambda x: x.replace('월',''))
````

````python
df.head()
````

# Q2-4

'지역' 컬럼을 시계열 데이타로 변경하고 인덱스로 설정하여라. 

````python
df['지역'] = pd.to_datetime(df['지역'])
````

````python
df.head()
````

# Q2-5

2015년 6월 이후의 데이타만 출력하여라 

````python
df[df['지역'] >= '2015-06-01']
````

# Q2-6

2014년도의 서울, 인천 경기 컬럼만 출력하여라. 

````python
df[['서울','인천','경기']][df['지역'].dt.year == 2014]
````

# Q2-7
csv 파일 형태로 현재의 데이타프레임을 저장하여라. 
<br>
output/평균분양가격_서울인천경기.csv

````python
df = df[['서울','인천','경기']][df['지역'].dt.year == 2014]
df.to_csv('평균분약가겨_서울인청경기.csv',encoding = 'cp949')
````
